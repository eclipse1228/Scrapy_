# Markov Text Generation (마르코프 모델) 
어떤 특정 사건이 다른 특정 사건에 뒤이어, 일정 확률로 일어나는 대규모 무작위 분포를 분석할 때 자주 쓰인다. 
## 마르코프 체인은 웹사이트들이 서로 어떻게 연결되는지를 모델화한다. (그래프로 표현 가능)


```python
from urllib.request import urlopen
from random import randint

def wordListSum(wordList):
    sum = 0
    for word, value in wordList.items():
        sum += value
    return sum

def retrieveRandomWord(wordList):
    randIndex = randint(1, wordListSum(wordList))
    for word, value in wordList.items():
        randIndex -= value
        if randIndex <= 0:
            return word

def buildWordDict(text):
    # 따옴표 제거
    text = text.replace('\n', ' ')
    text = text.replace('\"', '')
    # 구두점 역시 단어로 취급해서 마르코프 체인에 들어가게한다. 
    punctuation = [',', '.', ';', ':']
    for symbol in punctuation:
        text = text.replace(symbol, ' {} '.format(symbol));
    
    words = text.split(' ')
    # 빈 단어 제거
    words = [word for word in words if word != '']

    wordDict = {}
    for i in range(1, len(words)):
        if words[i-1] not in wordDict:
            # 새로운 단어를 위한 딕셔너리
            wordDict[words[i-1]] = {}
        if words[i] not in wordDict[words[i-1]]:
            wordDict[words[i-1]][words[i]] = 0
        wordDict[words[i-1]][words[i]] += 1
    return wordDict


speech = 'http://pythonscraping.com/files/inaugurationSpeech.txt'
content = str(urlopen(speech).read(), 'utf-8')
wordDict = buildWordDict(content)
# 길이 100인 마르코프 체인 생성
length = 100
chain = ['I']
for i in range(0, length):
    newWord = retrieveRandomWord(wordDict[chain[-1]])
    chain.append(newWord)

print(' '.join(chain))

# buildWordDict 함수는 텍스트를 받아서 마르코프 체인을 만든다. (ex) word_a : { word_b : 2 , word_c: 1}) 이면 1/3 확률로 word_b, 2/3 확률로 word_c가 나온다.는 의미이다! (그래프랑 똑같네)
# 계산은 행렬로 해도 되고, 딕셔너리로 해도 된다.(행렬로 하면 더 빠르다.)
# retrieveRandomWord 함수는 마르코프 체인을 이용해서 랜덤한 단어를 생성한다.
# chain은 마르코프 체인을 저장하는 리스트이다.


```

```python
def searchBreadth(targetPageId, paths=[[1]]):
    # paths는 현재까지의 경로를 저장한다.
    newPaths = []
    for path in paths:
        links = getLinks(path[-1])
        for link in links:
            if link == targetPageId:
                return path + [link]
            else:
                newPaths.append(path + [link])  
    return searchBreadth(targetPageId, newPaths)

def getLinks(pageId):
    html = urlopen('http://en.wikipedia.org/wiki/{}'.format(pageId))
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).findAll('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('Kevin_Bacon')
```
